```python
import csv
from datetime import datetime

EXPENSES_FILE = "expenses.csv"

# Load expenses from CSV file
def load_expenses():
    expenses = []
    try:
        with open(EXPENSES_FILE, mode='r', newline='') as file:
            reader = csv.DictReader(file)
            for row in reader:
                expenses.append({
                    'date': row['date'],
                    'category': row['category'],
                    'amount': float(row['amount']),
                    'description': row['description']
                })
    except FileNotFoundError:
        print("No previous expenses found. Starting fresh.")
    except Exception as e:
        print(f"Error loading expenses: {e}")
    return expenses

# Save expenses to CSV file
def save_expenses(expenses):
    try:
        with open(EXPENSES_FILE, mode='w', newline='') as file:
            fieldnames = ['date', 'category', 'amount', 'description']
            writer = csv.DictWriter(file, fieldnames=fieldnames)
            writer.writeheader()
            for exp in expenses:
                writer.writerow(exp)
        print("Expenses saved successfully.")
    except Exception as e:
        print(f"Error saving expenses: {e}")

# Show menu options
def show_menu():
    print("\n=== Expense Tracker Menu ===")
    print("1. Add New Expense")
    print("2. View All Expenses")
    print("3. View Expenses by Category")
    print("4. View Monthly Summary")
    print("5. Show Full Expense Summary")
    print("6. Save & Exit")

# Add new expense entry
def add_expense(expenses):
    try:
        date_str = input("Enter date (YYYY-MM-DD) or leave blank for today: ").strip()
        if date_str == '':
            date_str = datetime.today().strftime('%Y-%m-%d')
        else:
            datetime.strptime(date_str, '%Y-%m-%d')  # Validate

        category = input("Enter category (e.g., Food, Travel, Bills): ").strip()
        if not category:
            print("Category cannot be empty.")
            return

        amount = float(input("Enter amount: "))
        if amount <= 0:
            print("Amount must be positive.")
            return

        description = input("Enter description (optional): ").strip()

        expenses.append({
            'date': date_str,
            'category': category,
            'amount': amount,
            'description': description
        })
        print(f"✅ Expense added: {date_str} | {category} | ₹{amount:.2f} | {description}")
    except ValueError:
        print("❌ Invalid input. Please enter a correct date and numeric amount.")

# View all expenses
def view_expenses(expenses):
    if not expenses:
        print("No expenses to show.")
        return
    print(f"\n{'Date':<12}{'Category':<15}{'Amount':<10}{'Description'}")
    print("-" * 50)
    for exp in expenses:
        print(f"{exp['date']:<12}{exp['category']:<15}₹{exp['amount']:<9.2f}{exp['description']}")

# Filter expenses by category
def view_by_category(expenses):
    category = input("Enter category to filter by: ").strip()
    filtered = [e for e in expenses if e['category'].lower() == category.lower()]
    if not filtered:
        print(f"No expenses found for category '{category}'.")
        return
    print(f"\nExpenses in category '{category}':")
    view_expenses(filtered)

# Monthly summary with budget check
def view_monthly_summary(expenses, monthly_budget):
    try:
        month_input = input("Enter month to view (YYYY-MM) or leave blank for current month: ").strip()
        if not month_input:
            month_input = datetime.today().strftime('%Y-%m')

        monthly_expenses = [e for e in expenses if e['date'].startswith(month_input)]
        total = sum(e['amount'] for e in monthly_expenses)

        print(f"\n📆 Monthly Summary for {month_input}")
        print(f"Total Spent: ₹{total:.2f}")
        print(f"Budget Limit: ₹{monthly_budget:.2f}")
        if total > monthly_budget:
            print("⚠️ WARNING: You have exceeded your monthly budget!")
        else:
            print("✅ You are within your budget.")
    except Exception as e:
        print("Error while showing monthly summary:", e)

# Full summary of all expenses
def show_summary(expenses):
    if not expenses:
        print("No expenses to summarize.")
        return
    summary = {}
    total = 0.0
    for e in expenses:
        summary[e['category']] = summary.get(e['category'], 0) + e['amount']
        total += e['amount']
    print("\nExpense Summary by Category:")
    for cat, amt in summary.items():
        print(f"{cat:<15}: ₹{amt:.2f}")
    print(f"Total Expenses: ₹{total:.2f}")

# Main loop
def main():
    print("🧾 Welcome to Personal Expense Tracker 🧾")
    expenses = load_expenses()

    try:
        budget = float(input("Enter your monthly budget (₹): "))
    except ValueError:
        print("Invalid input. Default budget ₹10000 will be used.")
        budget = 10000

    while True:
        show_menu()
        choice = input("Enter your choice (1-6): ").strip()

        if choice == '1':
            add_expense(expenses)
        elif choice == '2':
            view_expenses(expenses)
        elif choice == '3':
            view_by_category(expenses)
        elif choice == '4':
            view_monthly_summary(expenses, budget)
        elif choice == '5':
            show_summary(expenses)
        elif choice == '6':
            save_expenses(expenses)
            print("Exiting Expense Tracker. Goodbye!")
            break
        else:
            print("❌ Invalid choice. Please select between 1 and 6.")

if __name__ == "__main__":
    main()
```
